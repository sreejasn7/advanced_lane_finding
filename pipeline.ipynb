```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import glob
import copy
%matplotlib inline

%matplotlib qt

ym_per_pix = 30/720
xm_per_pix = 3.7/700 
#img = cv2.imread('camera_cal/calibration1.jpg')
#plt.imshow(img)


def camera_calibrate(nx = 9 , ny=6):
    obj_points = []
    img_points = []
    
    objp = np.zeros((nx * ny , 3) , np.float32)
    objp[:,:2] = np.mgrid[0:nx,0:ny].T.reshape(-1,2)
    
    images = glob.glob('camera_cal/calibration*.jpg')
    for img_path in images:
        img = cv2.imread(img_path)
        gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        ret, corners = cv2.findChessboardCorners(gray, (nx,ny), None)
        if ret == True:
            obj_points.append(objp)
            img_points.append(corners)
            #img = cv2.drawChessboardCorners(img, (nx,ny), corners, ret)

            #plt.imshow(img)
    return obj_points , img_points
    #img_test = cv2.imread('camera_cal/calibration1.jpg')    
    #ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(obj_points, img_points, img_test.shape[1::-1], None, None)
    #dst = cv2.undistort(img_test, mtx, dist, None, mtx) 
    #cv2.imwrite('output_images/dis')
    #plt.imshow(dst)
    

def undistort_image(img, obj_points ,img_points):
    ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(obj_points, img_points, img.shape[1::-1], None, None)
    dst = cv2.undistort(img, mtx, dist, None, mtx) 
    return dst

def get_binary_image(img):
    sx_thresh=(20, 100)
    s_thresh=(170, 255)
    
    hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    l_channel = hls[:,:,1]
    s_channel = hls[:,:,2]
    #return s_channel
    sobelx = cv2.Sobel(l_channel, cv2.CV_64F, 1, 0)
    abs_sobelx = np.absolute(sobelx)
    scaled_sobel = np.uint8(255*abs_sobelx/np.max(abs_sobelx))
    
    sxbinary = np.zeros_like(scaled_sobel)
    sxbinary[(scaled_sobel >= sx_thresh[0]) & (scaled_sobel <= sx_thresh[1])] = 255
    
    s_binary = np.zeros_like(s_channel)
    s_binary[(s_channel >= s_thresh[0]) & (s_channel <= s_thresh[1])] = 255
    
    combined = np.bitwise_or(sxbinary,s_binary)
    return cv2.cvtColor(combined , cv2.COLOR_GRAY2RGB)
    
    #color_binary = np.dstack(( np.zeros_like(combined),combined, combined)) * 255
    #return color_binary
    


def perspective_transform(img):
    offset_leftright = 210
    offset_topbottom = 0
    img_size =  (img.shape[1], img.shape[0])
    #src = np.float32([[744,484] , [1050 , 672] , [256 , 672],[545 , 484]])
    src = np.float32([[744,484] , [1099 , 715] , [209 , 717],[545 , 484]])
    
    dst = np.float32([[img_size[0]-offset_leftright, offset_topbottom],[img_size[0]-offset_leftright, img_size[1]-offset_topbottom],\
                      [offset_leftright, img_size[1]-offset_topbottom],[offset_leftright, offset_topbottom]])
    
    M = cv2.getPerspectiveTransform(src, dst)
    wraped = cv2.warpPerspective(img, M, img_size , flags=cv2.INTER_LINEAR)
                     
    return wraped , src , dst


def inverseTransform(dst , src):
    Minv = cv2.getPerspectiveTransform(dst , src)
    return Minv
    
def find_lane_pixels(binary_warped):
    binary_warped2 = cv2.cvtColor(binary_warped , cv2.COLOR_RGB2GRAY)
    histogram = np.sum(binary_warped2[binary_warped2.shape[0]//2:,:], axis=0)
    #out_img = np.dstack((binary_warped, binary_warped, binary_warped))*255
    out_img = copy.deepcopy(binary_warped)
    
    #print(binary_warped.max() , binary_warped.mean() , binary_warped.dtype, binary_warped.shape)
    #out_img = cv2.cvtColor(binary_warped,cv2.COLOR_GRAY2RGB)
 
    midpoint = np.int(histogram.shape[0]//2)
    leftx_base = np.argmax(histogram[:midpoint])
    rightx_base = np.argmax(histogram[midpoint:]) + midpoint

    # HYPERPARAMETERS
    # Choose the number of sliding windows
    nwindows = 9
    margin = 100
    minpix = 50
    
    offset = midpoint  - ((leftx_base + rightx_base) / 2 )

    window_height = np.int(binary_warped2.shape[0]//nwindows)
    # Identify the x and y positions of all nonzero pixels in the image
    nonzero = binary_warped2.nonzero()
    nonzeroy = np.array(nonzero[0])
    nonzerox = np.array(nonzero[1])
    #print(nonzeroy)
    # Current positions to be updated later for each window in nwindows
    leftx_current = leftx_base
    rightx_current = rightx_base

    # Create empty lists to receive left and right lane pixel indices
    left_lane_inds = []
    right_lane_inds = []

    # Step through the windows one by one
    for window in range(nwindows):
        # Identify window boundaries in x and y (and right and left)
        win_y_low = binary_warped.shape[0] - (window+1)*window_height
        win_y_high = binary_warped.shape[0] - window*window_height
        ### TO-DO: Find the four below boundaries of the window ###
        win_xleft_low = leftx_current - margin  # Update this
        win_xleft_high = leftx_current + margin  # Update this
        win_xright_low =  rightx_current - margin # Update this
        win_xright_high = rightx_current + margin   # Update this
        
        # Draw the windows on the visualization image
        #print(win_xleft_low,win_y_low , win_xleft_high,win_y_high)
        cv2.rectangle(out_img,(win_xleft_low,win_y_low),(win_xleft_high,win_y_high),(0,255,0), 2) 
        cv2.rectangle(out_img,(win_xright_low,win_y_low),(win_xright_high,win_y_high),(0,255,0), 2) 
        
        ### TO-DO: Identify the nonzero pixels in x and y within the window ###
        good_left_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & \
                          (nonzerox >= win_xleft_low) & (nonzerox < win_xleft_high)).nonzero()[0]
        
        good_right_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & \
                           (nonzerox >= win_xright_low) &  (nonzerox < win_xright_high)).nonzero()[0]
        
        # Append these indices to the lists
        left_lane_inds.append(good_left_inds)
        right_lane_inds.append(good_right_inds)
       
    
        if len(good_left_inds) > minpix:
            leftx_current = np.int(np.mean(nonzerox[good_left_inds]))
        if len(good_right_inds) > minpix:        
            rightx_current = np.int(np.mean(nonzerox[good_right_inds]))


    # Concatenate the arrays of indices (previously was a list of lists of pixels)
    try:
        left_lane_inds = np.concatenate(left_lane_inds)
        right_lane_inds = np.concatenate(right_lane_inds)
    except ValueError:
        # Avoids an error if the above is not implemented fully
        pass

    # Extract left and right line pixel positions
    leftx = nonzerox[left_lane_inds]
    lefty = nonzeroy[left_lane_inds] 
    rightx = nonzerox[right_lane_inds]
    righty = nonzeroy[right_lane_inds]
    
    #plt.imshow(out_img)
    return leftx, lefty, rightx, righty, out_img ,offset

def fit_polynomial(binary_warped):
    leftx, lefty, rightx, righty, out_img , offset = find_lane_pixels(binary_warped)
    #print(leftx , lefty, rightx, righty)
    ### TO-DO: Fit a second order polynomial to each using `np.polyfit` ###
    global ym_per_pix , xm_per_pix
    
    left_fit = np.polyfit(lefty, leftx, 2)
    right_fit = np.polyfit(righty, rightx, 2)
    
    # Real time
    left_fit_real = np.polyfit(lefty * ym_per_pix, leftx * xm_per_pix, 2)
    right_fit_real = np.polyfit(righty * ym_per_pix, rightx * xm_per_pix, 2)
    

    # Generate x and y values for plotting
    ploty = np.linspace(0, binary_warped.shape[0]-1, binary_warped.shape[0] )
    try:
        left_fitx = left_fit[0]*ploty**2 + left_fit[1]*ploty + left_fit[2]
        right_fitx = right_fit[0]*ploty**2 + right_fit[1]*ploty + right_fit[2]
    except TypeError:
        # Avoids an error if `left` and `right_fit` are still none or incorrect
        print('The function failed to fit a line!')
        left_fitx = 1*ploty**2 + 1*ploty
        right_fitx = 1*ploty**2 + 1*ploty

    ## Visualization ##
    # Colors in the left and right lane regions
    out_img[lefty, leftx] = [255, 0, 0]
    out_img[righty, rightx] = [0, 0, 255]

    # DRAW--------------------
    
    left_p = np.int32(np.dstack((left_fitx , ploty)).reshape(-1,1,2))
    #print('yellow' , left_p)
    right_p = np.int32(np.dstack((right_fitx , ploty)).reshape(-1,1,2))
    
    out_img = cv2.polylines(out_img,[left_p],False,(255,255,0),thickness=15,lineType=cv2.LINE_AA)
    out_img = cv2.polylines(out_img,[right_p],False,(255,255,0),thickness=15,lineType=cv2.LINE_AA)
    

    return out_img , left_fitx  , right_fitx , ploty , offset , left_fit_real , right_fit_real

def measure_curvature_pixels(ploty, left_fit_r, right_fit_r,off_set):

    global ym_per_pix  , xm_per_pix 
    
    y_eval = np.max(ploty)
    calib_offset = off_set * xm_per_pix
    
    left_curverad = ((1 + (2*left_fit_r[0]*y_eval*ym_per_pix + left_fit_r[1])**2)**1.5) / np.absolute(2*left_fit_r[0])  ## Implement the calculation of the left line here
    right_curverad = ((1 + (2*right_fit_r[0]*y_eval*ym_per_pix + right_fit_r[1])**2)**1.5) / np.absolute(2*right_fit_r[0])  ## Implement the calculation of the right line here


    mean_c = np.mean([ left_curverad , right_curverad])
    
    return round(mean_c,2), left_curverad , right_curverad, round(calib_offset ,2)

def fill_final_frame(warped , undist_img , left_fitx , right_fitx,ploty,img_shape,Minv , mean_c,calib_offset):
    # Create an image to draw the lines on
    #warp_zero = np.zeros_like(warped).astype(np.uint8)
    #color_warp = np.dstack((warp_zero, warp_zero, warp_zero))
    color_warp = copy.deepcopy(warped)
    
    # Recast the x and y points into usable format for cv2.fillPoly()
    pts_left = np.array([np.transpose(np.vstack([left_fitx, ploty]))])
    pts_right = np.array([np.flipud(np.transpose(np.vstack([right_fitx, ploty])))])
    pts = np.hstack((pts_left, pts_right))
    # Draw the lane onto the warped blank image
    cv2.fillPoly(color_warp, np.int_([pts]), (0,255, 0))
    # Warp the blank back to original image space using inverse perspective matrix (Minv)
    newwarp = cv2.warpPerspective(color_warp, Minv, (img_shape[1], img_shape[0])) 
    # Combine the result with the original image
    result = cv2.addWeighted(undist_img, 1, newwarp, 0.3, 0)
    
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(result,'Radius Of Curvature = ' + str(mean_c) + '(m)' ,(64,100), font, 1 ,(255,255,255),3,cv2.LINE_AA)
    #print('Calib offset = ' , calib_offset)
    if (calib_offset > 0):
        pos = 'left'
    else:
        pos = 'right'
        
    #print(pos)
    #plt.imshow(result)
    cv2.putText(result,'Vehicle is ' + str(abs(calib_offset)) + 'm '+pos +' of center'  ,(64,200), font, 1,(255,255,255),3,cv2.LINE_AA)


    return result
    

def dummy(binary_warped):
    binary_wraped2 = cv2.cvtColor(binary_warped , cv2.COLOR_RGB2GRAY)
    histogram = np.sum(binary_wraped2[binary_wraped2.shape[0]//2:,:], axis=0)
    out_img = copy.deepcopy(binary_warped)
    
    #print(binary_warped.max() , binary_warped.mean() , binary_warped.dtype, binary_warped.shape)
    #out_img = cv2.cvtColor(binary_warped,cv2.COLOR_GRAY2RGB)
 
    midpoint = np.int(histogram.shape[0]//2)
    leftx_base = np.argmax(histogram[:midpoint])
    #plt.plot(histogram)

    rightx_base = np.argmax(histogram[midpoint:]) + midpoint
    
    #print('base points' , leftx_base , rightx_base,midpoint ,histogram[:midpoint].max())

    # HYPERPARAMETERS
    # Choose the number of sliding windows
    nwindows = 9
    margin = 100
    minpix = 50

    window_height = np.int(binary_warped.shape[0]//nwindows)
    # Identify the x and y positions of all nonzero pixels in the image
    nonzero = binary_warped.nonzero()
    nonzeroy = np.array(nonzero[0])
    nonzerox = np.array(nonzero[1])
    # Current positions to be updated later for each window in nwindows
    leftx_current = leftx_base
    rightx_current = rightx_base

    # Create empty lists to receive left and right lane pixel indices
    left_lane_inds = []
    right_lane_inds = []
    
    for window in range(nwindows):
        # Identify window boundaries in x and y (and right and left)
        win_y_low = binary_warped.shape[0] - (window+1)*window_height
        win_y_high = binary_warped.shape[0] - window*window_height
        ### TO-DO: Find the four below boundaries of the window ###
        win_xleft_low = leftx_current - margin  # Update this
        win_xleft_high = leftx_current + margin  # Update this
        win_xright_low =  rightx_current - margin # Update this
        win_xright_high = rightx_current + margin   # Update this
        
        # Draw the windows on the visualization image
        #print(win_xleft_low,win_y_low , win_xleft_high,win_y_high)
        cv2.rectangle(out_img,(win_xleft_low,win_y_low),(win_xleft_high,win_y_high),(0,255,0), 2) 
        cv2.rectangle(out_img,(win_xright_low,win_y_low),(win_xright_high,win_y_high),(0,255,0), 2)
        
        good_left_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & \
                          (nonzerox >= win_xleft_low) & (nonzerox < win_xleft_high)).nonzero()[0]
        
        good_right_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & \
                           (nonzerox >= win_xright_low) &  (nonzerox < win_xright_high)).nonzero()[0]
        
        if len(good_left_inds) > minpix:
            leftx_current = np.int(np.mean(nonzerox[good_left_inds]))
        if len(good_right_inds) > minpix:        
            rightx_current = np.int(np.mean(nonzerox[good_right_inds]))
        
    
    plt.imshow(out_img)

    
```
